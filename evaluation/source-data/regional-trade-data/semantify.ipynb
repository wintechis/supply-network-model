```python
import pandas as pd
df = pd.read_csv("EURegionalIOtable_2010.csv", sep=";")
df.head(6)
```

```python
# remove first two rows because they contain headings only
df.drop(df.index[:2], inplace=True)
# remove the first column because it contains headings only
df.drop(columns=df.columns[[0]], axis=1, inplace=True)
# remove the last 1071 columns because they contain other data
df.drop(columns=df.columns[-1071:], axis=1, inplace=True)
# remove additional 238 columns because they contain data on countrise or continents
df.drop(columns=df.columns[-238:], axis=1, inplace=True)
df.head(6)
```

```python
# show last rows
df.tail(5)
```

```python
# remove last 11 rows because they contain other data
df = df[:-11]
# remove additional 238 rows because they contain data on countrise or continents
df = df[:-238]
df.tail(5)
```

```python
from rdflib import Graph, Literal, BNode, Namespace
from rdflib.namespace import RDF, XSD
# Create an empty graph
g = Graph()
# Define namespaces
EX = Namespace("http://example.org/")
g.bind("ex", EX)
SN = Namespace("https://purl.org/sn/vocab#")
g.bind("sn", SN)
FSN = Namespace("https://purl.org/fsn/vocab#")
g.bind("fsn", FSN)
CSN = Namespace("https://example.org/csn/vocab#")
g.bind("csn", CSN)
ISN = Namespace("https://example.org/isn/vocab#")
g.bind("isn", ISN)
CUR = Namespace("http://qudt.org/vocab/currency/")
g.bind("cur", CUR)

# Output file
result_file = "streamed_output.nt"

# Clear result file by opening it in write mode
with open(result_file, "w", encoding="utf-8") as file:
    pass  # Opening in 'w' mode clears the file content

for index, row in df.iterrows():
    columnnumber = 5
    while columnnumber < len(df.columns) :
        columnnumber += 1
        columnname = "Unnamed: " + str(columnnumber)
        try:
            quantity = float(row[columnname].replace(',','.'))
        except:
            continue
        input_region = df.iloc[1,columnnumber-1]
        input_country = input_region[:-2]
        input_industry = df.iloc[3,columnnumber-1]
        output_region = df.iloc[index-2,1]
        output_industry = df.iloc[index-2,3]
        relation = BNode()
        volume = BNode()
        g.add((EX[input_region+input_industry], RDF.type, SN['MultidimensionalEntity']))
        g.add((EX[input_region+input_industry], SN['dimension'], EX[input_region]))
        g.add((EX[input_region], RDF.type, FSN['FederalState']))
        g.add((EX[input_region], FSN['inCountry'], EX[input_country]))
        g.add((EX[input_country], RDF.type, CSN['Country']))
        g.add((EX[input_region+input_industry], SN['dimension'], EX[input_industry]))
        g.add((EX[input_industry], RDF.type, ISN['Industry']))
        g.add((relation, RDF.type, SN['Relation']))
        g.add((relation, SN['from'], EX[input_region+input_industry]))
        g.add((relation, SN['into'], EX[output_region+output_industry]))
        g.add((volume, RDF.type, SN['Volume']))
        g.add((volume, SN['relation'], relation))
        g.add((volume, SN['time'], EX['2010']))
        g.add((volume, SN['unit'], CUR['MegaUSD']))
        g.add((volume, SN['quantity'], Literal(quantity,datatype=XSD.decimal)))

        # Append triples to result file
        with open(result_file, "a", encoding="utf-8") as file:
            file.write(g.serialize(format="nt"))
        g.remove((None, None, None))

    if index > 7:
        break
```
